```python
import matplotlib.pyplot as plt
x=["java","c++","c","python"]
y=[80,90,50,70]
plt.xlabel("language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)
plt.title("Popularity",fontsize=15)
c=["r","y","b","g"]
plt.bar(x,y,width=0.4,color=c)
plt.show() 
```

```python
import matplotlib.pyplot as plt
x=["java","c++","c","python"]
y=[80,90,50,70]
plt.xlabel("language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)
plt.title("Popularity",fontsize=15)
c=["r","y","b","g"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="black",label="Popularity")
plt.legend()#use align ="edge" for edge
plt.show() 
```

```python
import matplotlib.pyplot as plt
x=["java","c++","c","python"]
y=[80,90,50,70]
plt.xlabel("language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)
plt.title("Popularity",fontsize=15)
c=["r","y","b","g"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="black",linewidth=5,linestyle=":",alpha=0.5) #alpha is used to reduce color
plt.show() 
```

```python
 import matplotlib.pyplot as plt
Day=[1,2,3,4,5,6,7,8,9]
Number=[3,5,4,2,0,6,6,4,8]
color=["r","b","y","r","y","g","pink","g","b"]
sizes = [10,90,58,70,65,5,40,3,25]
plt.scatter(Day,Number,c=color,s=sizes,alpha=0.6)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["China","India","Rusia","UK","Norway"]
y=[80,85,85,90,95]
z=[20,30,40,50,60]
plt.title("Development")
plt.xlabel("Country")
plt.ylabel("percentage")
c=["y","g","b","pink","r"]
plt.bar(x,y,width=0.5,color="r",label="Development")
plt.bar(x,z,width=0.5,color="orange",label="Develpment1")
plt.show()



```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["China","India","Rusia","UK","Norway"]
y=[80,85,85,90,95]
z=[20,30,40,50,60]
width=0.2
p=np.arange(len(x))#to seperate both plots
p1=[j+width for j in p]
plt.title("Development")
plt.xlabel("Country")
plt.ylabel("percentage")

plt.bar(p,y,width,color="r",label="Development")
plt.bar(p1,z,width,color="orange",label="Develpment1")
plt.show()
```
